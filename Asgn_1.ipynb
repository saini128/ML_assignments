Question 1(d): Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
```

```python
most_freq_x=np.bincount(x).argmax()
most_freq_y=np.bincount(y).argmax()
index_x=np.where(x == most_freq_x)[0]
index_y=np.where(y == most_freq_y)[0]
print('Most freq in x is ',most_freq_x,' and index is ',index_x)
print('Most freq in y is ',most_freq_y,' and index is ',index_y)
```

Question 1(f): For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find

i. Sum of diagonal elements

ii. Eigen values of matrix

iii. Eigen vectors of matrix

iv. Inverse of matrix

v. Determinant of matrix


```python
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
```

```python
sum_diagonal = np.trace(n_array)
eigenvalues, eigenvectors = np.linalg.eig(n_array)
inverse_matrix = np.linalg.inv(n_array)
determinant = np.linalg.det(n_array)

print('diagnol sum is ',sum_diagonal)
print('eiganvalues are ',eigenvalues)
print('eigen vectors are \n', eigenvectors)
print('inverse of the matrix is \n', inverse_matrix)
print('determinant ', determinant)
```

Question 1(g): Multiply the following matrices and also find covariance between matrices using NumPy:

i. p = [[1, 2], [2, 3]]

  q = [[4, 5], [6, 7]]

ii. p = [[1, 2], [2, 3], [4, 5]]

  q = [[4, 5, 1], [6, 7, 2]]


```python
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
```

```python
prod_1=np.dot(p1,q1)
covar_1 = np.cov(p1.flatten(), q1.flatten())
prod_2=np.dot(p2,q2)
covar_2 = np.cov(p2.flatten(), q2.flatten())
print('product of matrix 1\n',prod_1)
print('product of matrix 2\n',prod_2)
print('covariance of matrix 1\n',covar_1)
print('covariance of matrix 2\n',covar_2)
```

Q4: (a) Write a function named img_to_array(path) that reads an image from a specified path and save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale images)

```python
from PIL import Image
```

```python
text_file_path=""
def img_to_array(path):
    global text_file_path
    with Image.open(path) as img:
        img_array = np.array(img)
        if len(img_array.shape) == 3:
            img_type = 'RGB'
        elif len(img_array.shape) == 2:
            img_type = 'Grayscale'
        else:
            raise ValueError("Unsupported image format")

        img_array_text = np.array2string(img_array, separator=',', threshold=np.inf)

        text_file_path = path.rsplit('.', 1)[0] + '_array.txt'
        with open(text_file_path, 'w') as file:
            file.write(f"Image Type: {img_type}\n")
            file.write("Array Data:\n")
            file.write(img_array_text)

        print(f"Image array saved to {text_file_path}")
```
